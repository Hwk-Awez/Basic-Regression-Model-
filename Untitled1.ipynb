```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/Student_data.csv")
```

```python
df
```

```python
plt.scatter(df['StudyHoursPerWeek'],df['Scores'])
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
```

```python
hours=df['StudyHoursPerWeek'].to_numpy()
scores=df['Scores'].to_numpy()

print(hours)
print(scores)
```

```python
m=0
c=0
#Learning rate
alpha=0.0001
n=len(hours)
epochs=1000
```

```python
for i in range(epochs):

  ynew=m*hours+c

  error=ynew-scores

  mtotal=0

  for j in range(n):
    mtotal+= hours[j] * error[j]

  mgrad= (1/n)*mtotal

  ctotal=0

  for k in range(n):
    ctotal+= error[k]

  cgrad=(1/n)*ctotal

  m-= alpha * mgrad
  c-= alpha * cgrad
```

```python
print("Final Value of m:",m)
print("Final Value of c:",c)
```

```python
mf=int(round(m,2))
cf=int(round(c,2))

plt.plot(hours,ynew)
plt.scatter(df['StudyHoursPerWeek'],df['Scores'])
plt.title("Comparison_Graph")
plt.xlabel("Hours")
plt.ylabel("Predicted_scores")
```

```python
testhr=35
print('Predicted Score:',round(m*testhr+c,2))
```
